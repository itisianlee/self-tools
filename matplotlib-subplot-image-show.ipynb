## 1. 多张图片显示

```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = plt.imread('./data/e3.jpg')
```

```python
row_num = 4
col_num = 4
plt.figure(figsize=(col_num, 4)) #设置窗口大小
plt.subplots_adjust(wspace=0, hspace=0) #调整子图间距
for idx in range(row_num*col_num):
    m = idx // row_num + 1
    n = idx % row_num + 1
    plt.subplot(row_num, col_num, idx+1)
    plt.imshow(img)
    plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(10,3)) #设置窗口大小
plt.suptitle('Multi_Image') # 图片名称
plt.subplots_adjust(wspace=0) #调整子图间距
plt.subplot(1,6,1), plt.title('im')
plt.imshow(img), plt.axis('off')

plt.subplot(1,6,2), plt.title('g')
plt.imshow(img), plt.axis('off')
plt.subplot(1,6,3), plt.title('i')
plt.imshow(img), plt.axis('off')
plt.subplot(1,6,4), plt.title('r')
plt.imshow(img), plt.axis('off')
plt.subplot(1,6,5), plt.title('g')
plt.imshow(img), plt.axis('off')
plt.subplot(1,6,6), plt.title('b')
plt.imshow(img), plt.axis('off')

plt.show()
```

## 3. axes坐标轴

```python
plt.figure(num=3, figsize=(8, 5))
plt.plot(x, y1)
plt.plot(x, y2, color='red', linestyle='--')
plt.xlim((-1, 2)) # 取值范围
plt.ylim((-2, 3))
plt.xlabel('i am X') # 对坐标轴取名字
plt.ylabel('i am Y')
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks) # 对坐标轴的单位进行改变
_ = plt.yticks([-2, -1.8, -1, 1.22, 3], 
           [r'$r\ bad$', r'$bad\ \alpha$', r'$normal$', r'$good$', r'$r\ good$']) 
```

- gca: 'get current axis'

```python
plt.figure(num=3, figsize=(8, 5))
plt.plot(x, y1)
plt.plot(x, y2, color='red', linestyle='--')
plt.xlim((-1, 2)) # 取值范围
plt.ylim((-2, 3))
plt.xlabel('i am X') # 对坐标轴取名字
plt.ylabel('i am Y')
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks) # 对坐标轴的单位进行改变
plt.yticks([-2, -1.8, -1, 1.22, 3], 
           [r'$r\ bad$', r'$bad\ \alpha$', r'$normal$', r'$good$', r'$r\ good$']) 
ax = plt.gca()
# 坐标轴的4个边框, ax.spines
ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none') 
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('axes', 0.3)) # outward, axes(使用的是百分比)
ax.spines['left'].set_position(('data', 0.5))
```

## 4. legend 图例(传奇,说明,图例,刻印文字)

```python
plt.figure(num=3, figsize=(8, 5))

plt.xlim((-1, 2)) # 取值范围
plt.ylim((-2, 3))
plt.xlabel('i am X') # 对坐标轴取名字
plt.ylabel('i am Y')
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks) # 对坐标轴的单位进行改变
plt.yticks([-2, -1.8, -1, 1.22, 3], 
           [r'$r\ bad$', r'$bad\ \alpha$', r'$normal$', r'$good$', r'$r\ good$']) 

l1, = plt.plot(x, y1, label='line1')
l2, = plt.plot(x, y2, color='red', linestyle='--', label='line2')
# 添加图例
_ = plt.legend(handles=[l1, l2], labels=['new l1', 'new l2'], loc='lower right')
```

## 5. annotation 添加注解

```python
plt.figure(num=3, figsize=(8, 5))

plt.xlim((-1, 2)) # 取值范围
plt.ylim((-2, 3))
plt.xlabel('i am X') # 对坐标轴取名字
plt.ylabel('i am Y')
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks) # 对坐标轴的单位进行改变
plt.yticks([-2, -1.8, -1, 1.22, 3], 
           [r'$r\ bad$', r'$bad\ \alpha$', r'$normal$', r'$good$', r'$r\ good$']) 

l1, = plt.plot(x, y1, label='line1', color='black')
l2, = plt.plot(x, y2, color='red', linestyle='--', label='line2')
# 添加图例
_ = plt.legend(handles=[l1, l2], labels=['new l1', 'new l2'], loc='lower right')
ax = plt.gca()
ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none') 
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0))
# 添加annotation
x0 = 0.5
y0 = 2 * x0 + 1
plt.scatter(x0, y0)
plt.plot([x0, x0], [y0, 0], 'k--', lw=2.5) # 打印出垂直与x轴的虚线
plt.annotate(r'$2x+1={}$'.format(y0), xy=(x0, y0), xycoords='data', 
             xytext=(+20, -20), textcoords='offset points', fontsize=10, 
             arrowprops=dict(arrowstyle='->'))
```

## 6. 散点图

```python
n = 1024
x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)
c = np.arctan2(y, x) # 颜色值

plt.scatter(x, y, s=75, c=c, alpha=0.5)
plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))

# 注释掉坐标轴的刻度
plt.xticks(())
_ = plt.yticks(())
```

## 7. 柱状图

```python
n = 12
x = np.arange(n)
y1 = (1-x/float(n)) * np.random.uniform(0.5, 1.0, n)
y2 = (1-x/float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(x, +y1, facecolor='r', edgecolor='white')
plt.bar(x, -y2, facecolor='b', edgecolor='white')

for x_, y_ in zip(x, y1):
    # ha: horizontal alignment
    plt.text(x_, y_+0.05, '{:.2f}'.format(y), ha='center', va='bottom')
    
for x_, y_ in zip(x, y2):
    # ha: horizontal alignment
    plt.text(x_, -y_-0.05, '{:-.2f}'.format(y), ha='center', va='top')
plt.xlim(-0.5, n)
plt.xticks(())
plt.ylim(-1.25, 1.25)
_ = plt.yticks(())
```
